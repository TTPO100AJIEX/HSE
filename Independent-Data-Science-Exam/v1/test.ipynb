```python
import sklearn.metrics
import sklearn.model_selection

numpy.random.seed(42)

train_x, test_x, train_y, test_y = sklearn.model_selection.train_test_split(tr_x1, tr_y, test_size = 0.25)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)

import torch
import torch.utils.data

torch.manual_seed(42)

train_ds = torch.utils.data.TensorDataset(torch.tensor(train_x, dtype=torch.float),
                                          torch.tensor(train_y, dtype=torch.float).reshape(-1, 1))
test_ds = torch.utils.data.TensorDataset(torch.tensor(test_x, dtype=torch.float),
                                         torch.tensor(test_y, dtype=torch.float).reshape(-1, 1))

train_dl = torch.utils.data.DataLoader(train_ds, batch_size = 64, shuffle = True)
test_dl = torch.utils.data.DataLoader(test_ds, batch_size = len(test_ds))

model = torch.nn.Sequential(
    torch.nn.Linear(train_x.shape[1], 8), torch.nn.GELU(),
    torch.nn.Dropout(0.1), torch.nn.Linear(8, 1),
)

optimizer = torch.optim.AdamW(
    params = model.parameters(),
    # lr = 1e-5
)

for _ in range(100):
    for batch_x, batch_y in train_dl:
        model.train()
        pred = model(batch_x)
        loss = torch.nn.functional.binary_cross_entropy_with_logits(pred, batch_y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    with torch.no_grad():
        model.eval()
        batch_x, batch_y = next(iter(test_dl))
        pred = model(batch_x)
        # print(torch.nn.functional.cross_entropy(pred, batch_y))
        if sklearn.metrics.roc_auc_score(batch_y, pred) > 0.75:
            print(_, sklearn.metrics.roc_auc_score(batch_y, pred))


final_x = torch.tensor(te_x1, dtype = torch.float)
pred = model(final_x)
numpy.savetxt("result.txt", pred.detach().numpy())
```
