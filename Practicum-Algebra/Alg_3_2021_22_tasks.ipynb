```python
import sympy
from sympy import linsolve, Matrix, S, Symbol, symbols, Eq, linear_eq_to_matrix, simplify, parse_expr
```

```python
class input_data:
  def __init__(this, data):
    this.input_data = data.split("\n")
    this.last_used_index = -1
  def get_input(this, question):
    this.last_used_index = this.last_used_index + 1
    if (this.last_used_index >= len(this.input_data)):
      this.input_data.append(input(question))
    else:
      print(f"""{question}{this.input_data[this.last_used_index]}""")
    return(this.input_data[this.last_used_index])
```

```python
def parse_equation(string, variables):
  string = string.replace("=", "-").replace(" ", "")
  koeffs = [0] * (len(variables) + 1)
  koeff = ""
  i = 0
  while (i < len(string)):
    checker = False
    for j in range(len(variables)):
      if (string[i:i + len(variables[j])] == variables[j]):
        if (koeff == "-"): koeff = "-1"
        if (koeff == "+"): koeff = "+1"
        if (koeff[len(koeff) - 1] == "*"): koeff = koeff[1:len(koeff) - 1]
        koeffs[j] += parse_expr(koeff)
        koeff = ""
        checker = True
        i = i + len(variables[j]) - 1
        break
    if (not checker):
      koeff += string[i]
    i += 1
  koeffs[len(variables)] = parse_expr(koeff)
  return(koeffs)
```

### Задание 1.
Решить СЛАУ, представив их тремя различными способами (расширенная матрица, список уравнений, матричный вид)
$$
a)\ \left\{
\begin{matrix}
-x+5y-3z=8\\
4x-y+5z=-1\\
3x+4y+5z=10
\end{matrix}
\right., \quad 
b)\ \left\{
\begin{matrix}
-x+5y-3z=8\\
4x-y+5z=-1\\
3x+4y+2z=7
\end{matrix}
\right., \quad 
c)\ \left\{
\begin{matrix}
-x+5y-3z=8\\
4x-y+5z=-1\\
3x+4y+2z=5
\end{matrix}
\right.
$$

```python
class SOLE:
  def __init__(this, variables_amount):
    this.variables = [None] * variables_amount

    this.left_side = []
    this.right_side = []

    this.augmented = []

    this.eq_list = []

    this.input_variables()
    this.input_equations()
  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = user_data.get_input(f"""Введите переменную {i + 1}: """)
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("Введите уравнение: ");
      koeffs = parse_equation(equation, this.variables)

      this.left_side.append([])
      for j in range(len(koeffs) - 1):
        this.left_side[len(this.left_side) - 1].append(koeffs[j])
      this.right_side.append(-koeffs[len(koeffs) - 1])

      koeffs[len(koeffs) - 1] = -koeffs[len(koeffs) - 1]
      this.augmented.append(koeffs)

      sum = 0
      for i in range(len(koeffs) - 1):
        sum += koeffs[i] * Symbol(this.variables[i])
      this.eq_list.append(Eq(sum, koeffs[len(koeffs) - 1]))
      
    this.left_side = Matrix(this.left_side)
    this.right_side = Matrix(this.right_side)
    
    this.augmented = Matrix(this.augmented)

    for i in range(len(this.variables)):
      this.variables[i] = Symbol(this.variables[i])

  def solve_augmented(this):
    return(linsolve(this.augmented))
  def solve_eq_list(this):
    return(linsolve(this.eq_list, *this.variables))
  def solve_matrix(this):
    return(linsolve((this.left_side, this.right_side)))
```

### Задание 2.

Решить СЛАУ с параметром $a$, представив их списком уравнений, а затем приведя к матричному виду с помощью linear_eq_to_matrix
$$
a)\ \left\{
\begin{matrix}
ax+5y-3z=8\\
4x-y+5z=-1\\
3x+4y+5z=10
\end{matrix}
\right., \quad 
b)\ \left\{
\begin{matrix}
(a-1)x+5y-3z=8\\
(4+a)x-y+5z=-1\\
(3+2a)x+4y+2z=7
\end{matrix}
\right., \quad 
c) \left\{
\begin{matrix}
-x+5y-3z=8a\\
4x-y+5z=-a\\
3x+4y+2z=5a
\end{matrix}
\right.
$$
Решить СЛАУ а) при $a=-82/25$, b) при $a=-19/6$, $-3.17$ и $-3.167$, c) при $a=1$.

Проанализировать результаты. Для каждого случая (кроме случая несовместной СЛАУ) провести проверку подстановкой.

При подстановке дробей пользоваться конструкцией вида $S(num)/den$, например,
$S(2)/3$

```python
class SOLE:
  def __init__(this, variables_amount):
    this.variables = [None] * variables_amount

    this.left_side = []
    this.right_side = []

    this.eq_list = []

    this.input_variables()
    this.input_equations()
  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = user_data.get_input(f"""Введите переменную {i + 1}: """)
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("Введите уравнение: ");
      koeffs = parse_equation(equation, this.variables)

      sum = 0
      for i in range(len(koeffs) - 1):
        sum += koeffs[i] * Symbol(this.variables[i])
      this.eq_list.append(Eq(sum, -koeffs[len(koeffs) - 1]))
      
    for i in range(len(this.variables)):
      this.variables[i] = Symbol(this.variables[i])

    this.left_side, this.right_side = linear_eq_to_matrix(this.eq_list, this.variables)

  def solve_matrix(this):
    return(linsolve((this.left_side, this.right_side)))
  def solve_subs(this, subs):
    return(linsolve((this.left_side.subs(subs), this.right_side.subs(subs))))
```

```python
user_data = input_data("""3
x
y
z
ax + 5y - 3z = 8
4x - y + 5z = -1
3x + 4y + 5z = 10
1
-82/25
3
x
y
z
(a - 1)x + 5y - 3z = 8
(4 + a)x - y + 5z = -1
(3 + 2 * a)x + 4y + 2z = 7
3
-19/6
-3.17
-3.167
3
x
y
z
-x + 5y - 3z = 8 * a
4x - y + 5z = -a
3x + 4y + 2z = 5 * a
1
1""")

for i in range(3):
  print(f"""\033[95m{chr(ord('a') + i)})\033[0m""")
  sole = input_SOLE()
  solution = sole.solve_matrix()
  display(solution)
  if (solution):
    display(simplify(sole.left_side * Matrix(*solution) - sole.right_side))
  try: 
    a_amount = int(user_data.get_input("Введите количество подстановок переменной a: "))
  except:
    print("\033[91mВведённые данные некорректны")
    sys.exit(0)
  for j in range(a_amount):
    a = user_data.get_input("Введите значение переменной a для подстановки: ")
    fraction_slash = a.find('/')
    if (fraction_slash == -1):
      try:
        solution = sole.solve_subs({Symbol('a'): float(a)})
        display(solution)
        if (solution):
          display(simplify(sole.left_side.subs({Symbol('a'): float(a)}) * Matrix(*solution) - sole.right_side.subs({Symbol('a'): float(a)})))
      except:
        print("\033[91mВведённые данные некорректны")
        sys.exit(0)
    else:
      try:
        numerator = int(a[0:fraction_slash])
        denominator = int(a[fraction_slash + 1:len(a)])
        solution = sole.solve_subs({Symbol('a'): S(numerator) / denominator})
        display(solution)
        if (solution):
          display(simplify(sole.left_side.subs({Symbol('a'): S(numerator) / denominator}) * Matrix(*solution) - sole.right_side.subs({Symbol('a'): S(numerator) / denominator})))
      except:
        print("\033[91mВведённые данные некорректны")
        sys.exit(0)
```

### Задание 3
В расширенной матрице СЛАУ из Задания 1 а) 

1) выделить матрицу левой части

2) выделить столбец правой части

3) транспонировать матрицу, полученную в 1)

4) получить расширенную матрицу из матрицы 3) и столбца 2)

5) создать на основе лямбда-функции матрицу $3\times 4$ из расположенных в шахматном порядке чисел 1 и $-1$.

```python
class SOLE:
  def __init__(this, variables_amount):
    this.variables = [None] * variables_amount

    this.left_side = []
    this.right_side = []

    this.augmented = []

    this.eq_list = []

    this.input_variables()
    this.input_equations()
  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = user_data.get_input(f"""Введите переменную {i + 1}: """)
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("Введите уравнение: ");
      koeffs = parse_equation(equation, this.variables)

      this.left_side.append([])
      for j in range(len(koeffs) - 1):
        this.left_side[len(this.left_side) - 1].append(koeffs[j])
      this.right_side.append(-koeffs[len(koeffs) - 1])

      koeffs[len(koeffs) - 1] = -koeffs[len(koeffs) - 1]
      this.augmented.append(koeffs)

      sum = 0
      for i in range(len(koeffs) - 1):
        sum += koeffs[i] * Symbol(this.variables[i])
      this.eq_list.append(Eq(sum, koeffs[len(koeffs) - 1]))
      
    this.left_side = Matrix(this.left_side)
    this.right_side = Matrix(this.right_side)
    
    this.augmented = Matrix(this.augmented)

    for i in range(len(this.variables)):
      this.variables[i] = Symbol(this.variables[i])

  def solve_augmented(this):
    return(linsolve(this.augmented))
  def solve_eq_list(this):
    return(linsolve(this.eq_list, *this.variables))
  def solve_matrix(this):
    return(linsolve((this.left_side, this.right_side)))
```

```python
user_data = input_data("""3
x
y
z
-x + 5y - 3z = 8
4x - y + 5z = -1
3x + 4y + 5 * z = 10""")
sole = input_SOLE()
print(f"""\033[95mРасширенная матрица СЛАУ из 1 а)\033[0m""")
A = sole.augmented
display(A)
print(f"""\033[95m1) Матрица левой части\033[0m""")
ans1 = A.extract([i for i in range(A.rows)], [i for i in range(A.cols - 1)])
display(ans1)
print(f"""\033[95m2) Матрица правой части\033[0m""")
ans2 = A.extract([i for i in range(A.rows)], [A.cols - 1])
display(ans2)
print(f"""\033[95m3) Транспонированная матрица левой части\033[0m""")
ans3 = ans1.transpose()
display(ans3)
print(f"""\033[95m4) Расширенная матрица из 3) и 2)\033[0m""")
ans4 = ans3.row_join(ans2)
display(ans4)
print(f"""\033[95m5) Матрица 3x4 из расположенных в шахматном порядке 1 и -1\033[0m""")
display(Matrix(3, 4, lambda i, j: (-1) ** (i + j)))
print("\033[95mИли\033[0m")
display(Matrix(3, 4, lambda i, j: (-1) ** (i + j + 1)))
```

Затем составить из матрицы левой части и столбца правой расширенную матрицу СЛАУ и решить первым способом. После этого провести проверку подстановкой.

 A = \left[\begin{matrix}1 & -7 & 6 & 7\\-7 & -1 & 6 & 9\\\mu & 1 & -3 & 1\\8 & -6 & 0 & -2\end{matrix}\right],

```python
class SOLE:
  def __init__(this, variables_amount):
    this.variables = [None] * variables_amount

    this.left_side = []
    this.right_side = []

    this.augmented = []

    this.eq_list = []

    this.input_variables()
    this.input_equations()
  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = Symbol(f"""x{i}""")
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("Введите коэффициенты уравнения через один пробел: ");
      koeffs = equation.split(' ')
      expr = 0
      try:
        for i in range(len(this.variables)):
          expr += parse_expr(koeffs[i]) * this.variables[i]
        this.eq_list.append(Eq(expr, parse_expr(koeffs[len(koeffs) - 1])))
      except:
        print("\033[91mВведённые данные некорректны")
        sys.exit(0)
  def create_matrix(this):
    this.left_side, this.right_side = linear_eq_to_matrix(this.eq_list, this.variables)
  def create_augmented(this):
    this.augmented = this.left_side.row_join(this.right_side)

  def solve_augmented(this):
    return(linsolve(this.augmented))
  def solve_eq_list(this):
    return(linsolve(this.eq_list, *this.variables))
  def solve_matrix(this):
    return(linsolve((this.left_side, this.right_side)))
```
