```python
#https://www.sympy.org/en/index.html
# sympy - модуль для символьных преобразований
from sympy import *
```

```python
class input_data:
  def __init__(this, data):
    this.input_data = data.split("\n")
    this.last_used_index = -1
  def get_input(this, question):
    this.last_used_index = this.last_used_index + 1
    if (this.last_used_index >= len(this.input_data)):
      this.input_data.append(input(question))
    return(this.input_data[this.last_used_index])
```

```python
def parse_symbols(s): #parse_expr
  s = s.replace(" ", "").replace("-", " - ").replace("+", " + ").replace("*", " * ").replace("/", " / ")
  if (len(s) == 0):
    s = "1";
  if (s[0] == ' '):
    s = "0" + s
  s = s.split(' ')
  for i in range(len(s) - 1):
    if (s[i] != '+' and s[i] != '/' and s[i] != '*' and s[i] != '-' and s[i + 1] != '+' and s[i + 1] != '/' and s[i + 1] != '*' and s[i + 1] != '-'):
      s[i] += s[i + 1]
      s.pop(i + 1)
  for i in range(len(s)):
    if (s[i] != '+' and s[i] != '/' and s[i] != '-' and s[i] != '*'):
      if (s[i].isnumeric()):
        s[i] = int(s[i])
      else:
        s[i] = Symbol(s[i])
  i = 0
  while (i < len(s)):
    if (s[i] == '*'):
      s[i - 1] = s[i - 1] * s[i + 1]
    if (s[i] == '/'):
      s[i - 1] = s[i - 1] / s[i + 1]
    if (s[i] == '*' or s[i] == '/'):
      s.pop(i)
      s.pop(i)
      i = i - 1
    i = i + 1
  i = 0
  while (i < len(s)):
    if (s[i] == '+'):
      s[i - 1] = s[i - 1] + s[i + 1]
    if (s[i] == '-'):
      s[i - 1] = s[i - 1] - s[i + 1]
    if (s[i] == '+' or s[i] == '-'):
      s.pop(i)
      s.pop(i)
      i = i - 1
    i = i + 1
  return(s[0])
```

```python
import sys

class SOLE:
  def __init__(this, variables_amount):
    this.equations = [[0] * variables_amount for i in range(variables_amount)]
    this.results = [0] * variables_amount
    this.variables = [None] * variables_amount
    this.input_variables()
    this.input_equations()
  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = user_data.get_input(f"""Введите переменную {i + 1}: """)
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("Введите уравнение: ");
      equation = equation.replace("*", "").replace("=", "-").replace(" ", "").replace("+-", "-").replace("--", "+").replace("++", "+").replace("-", " -").replace("+", " +")
      equation = equation.split(' ')
      for term in equation:
        if (len(term) == 0):
          continue
        found_variable = false
        for j in range(len(this.variables)):
          if (this.variables[j] in term):
            term = term.replace(this.variables[j], "")
            if (term == '+'):
              term = "1"
            elif (term == '-'):
              term = '-1'
            this.equations[i][j] += parse_symbols(term)
            found_variable = true
            break
        if (not found_variable):
          this.results[i] -= parse_symbols(term)
    this.equations = Matrix(this.equations)
    this.results = Matrix(this.results)


  def solve(this):
    det_equations = this.equations.det()
    this.solutions = [None] * len(this.variables)
    for i in range(len(this.variables)):
      __column = this.equations.col(i)
      this.equations.col_del(i)
      this.equations = this.equations.col_insert(i, this.results)
      this.solutions[i] = this.equations.det() / det_equations
      this.equations.col_del(i)
      this.equations = this.equations.col_insert(i, __column)
    this.solutions = Matrix(this.solutions)
    return(this.solutions)
```

```python
def input_SOLE():
  try: 
    variables_amount = int(user_data.get_input("Введите количество переменных:"))
  except:
    print("\033[91mВведённые данные некорректны")
    sys.exit(0)
  return(SOLE(variables_amount))
```

### Задание 3.
Решить по формулам Крамера СЛАУ с параметрами
\begin{align*}
\left\{\begin{matrix} x_{1} + a_{12} x_{2}  = b_1
\\ x_{1} + a_{22} x_{2}  = b_2
\end{matrix}\right.
\end{align*}
Проверить правильность найденного решения подстановкой.

Найти такие наборы параметров, для которых полученное решение неприменимо (подсказка - из-за деления на ноль).

Показать, что для этих наборов определитель матрицы левой части равен нулю, использовать подстановки subs.

```python
user_data = input_data("""2
x1
x2
x1 + a12x2 = b1
x1 + a22x2 = b2""")

system = input_SOLE()

system.solve()
for i in range(len(system.variables)):
  print(f"""{system.variables[i]} = {system.solutions[i]}""")
print("\n")
display(system.solutions)
print("\n")
display(simplify(system.equations * system.solutions - system.results))

print(f"""\n\033[95mРешение неприменимо, когда определитель левой части СЛАУ равен 0:\033[0m""")
print(f"""\033[4mМатрица левой части СЛАУ:""")
display(system.equations)
print(f"""\033[4mОпределитель левой части СЛАУ:""")
display(system.equations.det())
print(f"""\033[4mНуль определителя:""")
solutions = solve(system.equations.det())
display(solutions)
for substitution in solutions:
  for key in substitution:
    system.equations = system.equations.subs(key, substitution[key])
print(f"""\033[4mМатрица левой части СЛАУ с подставленными решениями:""")
display(system.equations)
print(f"""\033[4mОпределитель матрицы левой части СЛАУ с подставленными решениями:""")
display(system.equations.det())
```

### Задание 4.
Дана матрица 
\begin{align*}
M =
\left(\begin{matrix} 
p & 1 & 0 \\
1 & q & 1 \\
0 & 1 & r
\end{matrix}\right)
\end{align*}
Вычислить определитель этой матрицы.
Найти целочисленные наборы параметров $p$, $q$, $r$, для которых определитель $M$ равен нулю. Рассматривать целые числа от 0 до 12 (НЕ включительно). Использовать подстановки subs в заранее вычисленный определитель.

Из найденных наборов как столбцов составить матрицу.

```python
p, q, r = symbols('p q r')
M = Matrix([[p, 1, 0], [1, q, 1], [0, 1, r]])
print(f"""\033[95mМатрица:""")
display(M)
print(f"""\033[95mОпределитель матрицы:""")
det = M.det()
display(det)

answer = [[], [], []]
for p_value in range(0, 12):
  for q_value in range(0, 12):
    for r_value in range(0, 12):
      det_value = det.subs({p: p_value, q: q_value, r: r_value})
      if (det_value == 0):
        answer[0].append(p_value)
        answer[1].append(q_value)
        answer[2].append(r_value)
display(Matrix(answer))
```
