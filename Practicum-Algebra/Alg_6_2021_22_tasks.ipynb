```python
import sympy
from sympy import  Symbol, symbols, S, Point, Line, Plane, Point3D, Matrix
```

```python
class input_data:
  def __init__(this, data):
    this.input_data = data.split("\n")
    this.last_used_index = -1
  def get_input(this, question):
    this.last_used_index = this.last_used_index + 1
    if (this.last_used_index >= len(this.input_data)):
      this.input_data.append(input(question))
    else:
      print(f"""{question}{this.input_data[this.last_used_index]}""")
    return(this.input_data[this.last_used_index])
```

```python
def input_points():
  points = {}
  points_amount = int(user_data.get_input("Введите количество точек: "))
  for i in range(points_amount):
    variable = user_data.get_input("Введите название и координаты переменной: ").split(' ')
    points[variable[0]] = Point(variable[1], variable[2], variable[3])
  return(points)
```

###  Задание 6*.
Даны точки  пространстве
$A(-2, 7, -5)$, $B(6, 2, -4)$, $C(14, -3, -2)$, $D(22, -8, -2)$, $K(30, -7, -4)$,
$M(94, -35, -2)$.

Определить, какие тройки точек лежат на одной прямой, какие четверки - в одной плоскости.

Для проверки того, что три точки лежат на одной прямой использовать Point3D.are_collinear.

Вывести на экран тройки точек, лежащие на одной прямой и четверки точек, лежащих в одной плоскости.

```python
from itertools import dropwhile

user_data = input_data("""6
A -2 7 -5
B 6 2 -4
C 14 -3 -2
D 22 -8 -2
K 30 -7 -4
M 94 -35 -2""")

points = input_points()

print("\033[95mНа одной прямой:\033[0m")
for key1 in sorted(points):
  for key2 in dropwhile(lambda el: el <= key1, sorted(points)):
    for key3 in dropwhile(lambda el: el <= key2, sorted(points)):
      if (Point3D.are_collinear(points[key1], points[key2], points[key3])):
        print(f"""{key1} {key2} {key3}""")

print("\033[95mВ одной плоскости:\033[0m")
for key1 in points:
  for key2 in dropwhile(lambda el: el <= key1, sorted(points)):
    for key3 in dropwhile(lambda el: el <= key2, sorted(points)):
      for key4 in dropwhile(lambda el: el <= key3, sorted(points)):
        if (Point3D.are_coplanar(points[key1], points[key2], points[key3], points[key4])):
          print(f"""{key1} {key2} {key3} {key4}""")
```

###  Задание 7*.
Составить уравнения всех (различных!) плоскостей, определяемых точками задания 5.

Указание: решать аналогично Примеру 11, проверить, что $A$, $B$, $C$ не лежат на одной прямой и составить плоскость $ABC$, включить ее в список плоскостей. 

```python
user_data = input_data("""4
A -1 2 -4
B 2 -2 1
C 5 3 -6
D 0 3 -6""")
from itertools import dropwhile
points = input_points()

planes = []
for key1 in sorted(points):
  for key2 in dropwhile(lambda el: el <= key1, sorted(points)):
    for key3 in dropwhile(lambda el: el <= key2, sorted(points)):
      if (not Point3D.are_collinear(points[key1], points[key2], points[key3])):
        new_plane = Plane(points[key1], points[key2], points[key3])
        if (all(not plane.equals(new_plane) for plane in planes)):
          planes.append(new_plane)

def gcd(a, b):
  while (a != 0 and b != 0):
    if (a > b):
      a = a % b
    else:
      b = b % a
  return(a + b)
for plane in planes:
  coeffs = sympy.Poly(plane.equation()).coeffs()
  g = 0
  for coeff in coeffs:
    g = gcd(g, abs(coeff))
  display(sympy.Eq(plane.equation() / g, 0))
```
