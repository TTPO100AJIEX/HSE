### –ó–∞–¥–∞–Ω–∏–µ 1.
–û–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è –ª–∏ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ (–≤—Å–µ —Å—Ä–∞–∑—É). –ï—Å–ª–∏ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è, –Ω–∞–π—Ç–∏ —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è.

–°–æ—Å—Ç–∞–≤–∏—Ç—å –°–õ–ê–£ –≤ –≤–∏–¥–µ —Å–ø–∏—Å–∫–∞ —É—Ä–∞–≤–Ω–µ–Ω–∏–π, —Ä–µ—à–∏—Ç—å —Å –ø–æ–º–æ—â—å—é linsolve.

–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–æ–π –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≤ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –ø–ª–æ—Å–∫–æ—Å—Ç–µ–π, —á—Ç–æ –Ω–∞–π–¥–µ–Ω–Ω–∞—è —Ç–æ—á–∫–∞ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –∫–∞–∂–¥–æ–π –∏–∑ –Ω–∏—Ö.

$-8x+5y-2z+28=0$

$-11x-10y-16z+71=0$

$-7x+3y+8z-27=0$.

```python
def parse_equation(string, variables):
  string = string.replace("=", "-").replace(" ", "")
  string = "+" + string
  koeffs = [0] * (len(variables) + 1)
  koeff = ""
  i = 0
  while (i < len(string)):
    checker = False
    for j in range(len(variables)):
      if (string[i:i + len(variables[j])] == variables[j]):
        if (koeff == "-"): koeff = "-1"
        if (koeff == "+"): koeff = "+1"
        if (koeff[len(koeff) - 1] == "*"): koeff = koeff[1:len(koeff) - 1]
        koeffs[j] += parse_expr(koeff)
        koeff = ""
        checker = True
        i = i + len(variables[j]) - 1
        break
    if (not checker):
      koeff += string[i]
    i += 1
  koeffs[len(variables)] = parse_expr(koeff)
  return(koeffs)
```

```python
class SOLE:
  def __init__(this, variables_amount):
    this.variables = [None] * variables_amount
    this.eq_list = []
    this.input_variables()
    this.input_equations()

  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = user_data.get_input(f"""–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é {i + 1}: """)
      
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("–í–≤–µ–¥–∏—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ: ");
      koeffs = parse_equation(equation, this.variables)

      sum = 0
      for i in range(len(koeffs) - 1):
        sum += koeffs[i] * Symbol(this.variables[i])
      this.eq_list.append(Eq(sum, -koeffs[len(koeffs) - 1]))

    for i in range(len(this.variables)):
      this.variables[i] = Symbol(this.variables[i])

  def solve_eq_list(this):
    return(linsolve(this.eq_list, *this.variables))
```

```python
user_data = input_data("""3
x
y
z
-8x + 5y - 2z + 28 = 0
-11x - 10y - 16z + 71 = 0
-7x + 3y + 8z - 27 = 0
""")
sole = input_SOLE()
solution = sole.solve_eq_list()
display(solution)
solution = tuple(*solution)
for i in range(len(sole.eq_list)):
  substitution = {}
  for j in range(len(sole.variables)):
    substitution[sole.variables[j]] = solution[j]
  display(sole.eq_list[i].subs(substitution))
```

### –ó–∞–¥–∞–Ω–∏–µ 2.
–ù–∞–π—Ç–∏ —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –ø–ª–æ—Å–∫–æ—Å—Ç–∏, –≤ –∫–æ—Ç–æ—Ä–æ–π –ª–µ–∂–∞—Ç —Ç–æ—á–∫–∏ 
$K(4,3,0)$, $M(2,4,-2)$ –∏ $N(-2,2,-1)$ —Å –ø—Ä—è–º–æ–π, –ø—Ä–æ—Ö–æ–¥—è—â–µ–π —á–µ—Ä–µ–∑ —Ç–æ—á–∫–∏ $A(-3,5,1)$ –∏ $B(5,2,-3)$.

```python
class point:
  def __init__(this, x, y, z):
    this.x = x
    this.y = y
    this.z = z
```

```python
class plane:
  def read_point(this):
    p = user_data.get_input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–∫–∏ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split(' ')
    for i in range(len(p)):
      p[i] = parse_expr(p[i])
    return(point(p[0], p[1], p[2]))
  def __init__(this):
    p1 = this.read_point()
    p2 = this.read_point()
    p3 = this.read_point()
    p4 = point(Symbol('x'), Symbol('y'), Symbol('z'))
    this.equation = []
    for item in tuple((p4, p2, p3)):
      this.equation.append([item.x - p1.x, item.y - p1.y, item.z - p1.z])
    this.equation = Eq(Matrix(this.equation).det(), 0)
  def intersect_wth_line(this, line):
    ans = Matrix([*linsolve([this.equation] + line.equations, Symbol('x'), Symbol('y'), Symbol('z'), Symbol('t'))])
    ans.col_del(-1)
    return(ans)
```

```python
class line:
  def read_point(this):
    p = user_data.get_input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–∫–∏ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split(' ')
    for i in range(len(p)):
      p[i] = parse_expr(p[i])
    return(point(p[0], p[1], p[2]))
  def __init__(this):
    p1 = this.read_point()
    p2 = this.read_point()
    vec1 = Matrix([p2.x, p2.y, p2.z]) - Matrix([p1.x, p1.y, p1.z])
    eqs = Matrix([p1.x, p1.y, p1.z]) + Symbol('t') * vec1
    this.equations = []
    this.equations.append(Eq(Symbol('x'), eqs[0]))
    this.equations.append(Eq(Symbol('y'), eqs[1]))
    this.equations.append(Eq(Symbol('z'), eqs[2]))
  def intersect_wth_plane(this, plane):
    ans = Matrix([*linsolve(this.equations + [plane.equation], Symbol('x'), Symbol('y'), Symbol('z'), Symbol('t'))])
    ans.col_del(-1)
    return(ans)
```

```python
user_data = input_data("""4 3 0
2 4 -2
-2 2 -1
-3 5 1
5 2 -3""")

p = plane()
l = line()
display(p.intersect_wth_line(l))
print("\n")
display(l.intersect_wth_plane(p))
```

### –ó–∞–¥–∞–Ω–∏–µ 3.
–°–æ—Å—Ç–∞–≤–∏—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –ø—Ä—è–º–æ–π, —è–≤–ª—è—é—â–µ–π—Å—è –ª–∏–Ω–∏–µ–π –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –ø–ª–æ—Å–∫–æ—Å—Ç–µ–π $ùë¶+2ùëß‚àí4=0$ –∏ $4ùë•+3ùë¶+2ùëß‚àí4=0$. 

–£–∫–∞–∑–∞–Ω–∏–µ: —Å–æ—Å—Ç–∞–≤–∏—Ç—å –°–õ–ê–£ –∏–∑ —É—Ä–∞–≤–Ω–µ–Ω–∏–π –ø–ª–æ—Å–∫–æ—Å—Ç–µ–π, –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è linsolve(), –∑–∞–º–µ–Ω–∏—Ç—å –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –≤ –ø–æ–ª—É—á–µ–Ω–Ω–æ–º —Ä–µ—à–µ–Ω–∏–∏ –Ω–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä t –∏ —Å–æ—Å—Ç–∞–≤–∏—Ç—å –°–õ–ê–£ –∏–∑ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π.

```python
class line:
  def read_plane(this):
    equation = user_data.get_input("–í–≤–µ–¥–∏—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ: ")
    koeffs = parse_equation(equation, ["x", "y", "z"])
    sum = 0
    for i in range(len(koeffs) - 1):
      sum += koeffs[i] * Symbol(this.variables[i])
    return(Eq(sum, -koeffs[len(koeffs) - 1]))
  def __init__(this):
    this.variables = ["x", "y", "z"]
    this.equations = []
    this.equations.append(this.read_plane())
    this.equations.append(this.read_plane())
    for i in range(len(this.variables)):
      this.variables[i] = Symbol(this.variables[i])
    this.equations = linsolve(this.equations, *this.variables)
    this.equations = Matrix(*this.equations).subs(Symbol('z'), Symbol('t'))
    this.equations = Eq(Matrix([*this.variables]), this.equations)
```

```python
user_data = input_data("""y + 2z - 4 = 0
4x + 3y + 2z - 4 = 0""")

l = line()
display(l.equations)
```

–£–∫–∞–∑–∞–Ω–∏–µ: —Å–æ—Å—Ç–∞–≤–∏—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –ø—Ä—è–º–æ–π –≤ –≤–∏–¥–µ —Å–ø–∏—Å–∫–∞ –ø—Ä–∞–≤—ã—Ö —á–∞—Å—Ç–µ–π —É—Ä–∞–≤–Ω–µ–Ω–∏–π, –∑–∞—Ç–µ–º –ø–æ –æ–±—Ä–∞–∑—Ü—É –ü—Ä–∏–º–µ—Ä–∞ 5 —Å–æ—Å—Ç–∞–≤–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–ª–æ—Å–∫–æ—Å—Ç–∏, –ø—Ä–æ—Ö–æ–¥—è—â–µ–π —á–µ—Ä–µ–∑ –Ω–∞—á–∞–ª–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç (–ø–æ —Ç—Ä–µ–º —Ç–æ—á–∫–∞–º). –î–ª—è —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–∏—è –≤—Ç–æ—Ä–æ–π –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –∑–∞–º–µ—Ç–∏–º, —á—Ç–æ –≤–µ–∫—Ç–æ—Ä –Ω–æ—Ä–º–∞–ª–∏ –∫ –ø–µ—Ä–≤–æ–π –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –∏ –ª—é–±–æ–π –≤–µ–∫—Ç–æ—Ä –ø—Ä—è–º–æ–π –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã –≤—Ç–æ—Ä–æ–π –ø–ª–æ—Å–∫–æ—Å—Ç–∏, –ø–æ—ç—Ç–æ–º—É –º–æ–∂–Ω–æ –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ñ–æ—Ä–º—É–ª—É 4 —É—Ä–∞–≤–Ω–µ–Ω–∏—è –ø–ª–æ—Å–∫–æ—Å—Ç–∏.
–í–µ–∫—Ç–æ—Ä –Ω–æ—Ä–º–∞–ª–∏ –º–æ–∂–Ω–æ —Å–æ—Å—Ç–∞–≤–∏—Ç—å –≤—Ä—É—á–Ω—É—é, –≥–ª—è–¥—è –Ω–∞ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –ø–ª–æ—Å–∫–æ—Å—Ç–∏.

```python
class plane:
  def __init__(this, construct, p1, p2, p3):
    if (construct == "points"):
      p4 = point(Symbol('x'), Symbol('y'), Symbol('z'))
      this.equation = []
      for item in tuple((p4, p2, p3)):
        this.equation.append([item.x - p1.x, item.y - p1.y, item.z - p1.z])
      this.equation = Eq(Matrix(this.equation).det(), 0)
    if (construct == "parallel_vectors"):
        this.equation = Eq(Matrix([
            [Symbol('x') - p1.x, Symbol('y') - p1.y, Symbol('z') - p1.z],
            [p2.x, p2.y, p2.z],
            [p3.x, p3.y, p3.z]
        ]).det(), 0)
```

```python
class line:
  def __init__(this):
    this.equations = Eq(
        Matrix([Symbol('x'), Symbol('y'), Symbol('z')]),
        Matrix([
              parse_expr(user_data.get_input("–í–≤–µ–¥–∏—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ: ")),
              parse_expr(user_data.get_input("–í–≤–µ–¥–∏—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ: ")),
              parse_expr(user_data.get_input("–í–≤–µ–¥–∏—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ: "))
        ]),
        evaluate=False
    )
  def create_plane(this, p):
    p1 = this.equations.subs(Symbol('t'), 0)
    p1 = point(p1.rhs[0], p1.rhs[1], p1.rhs[2])
    p2 = this.equations.subs(Symbol('t'), 1)
    p2 = point(p2.rhs[0], p2.rhs[1], p2.rhs[2])
    return(plane("points", p1, p2, p))
```

```python
user_data = input_data("""2 + 5*t
-1 - 2*t
4 - 3*t""")

l = line()
p1 = l.create_plane(point(0, 0, 0))

display()

koeffs = sympy.Poly(p1.equation, Symbol('x'), Symbol('y'), Symbol('z')).coeffs()
point1 = l.equations.subs(Symbol('t'), 0).rhs
point2 = l.equations.subs(Symbol('t'), 1).rhs - point1
p2 = plane("parallel_vectors", point(point1[0], point1[1], point1[2]), point(koeffs[0], koeffs[1], koeffs[2]), point(point2[0], point2[1], point2[2]))
display(p1.equation)
display(p2.equation)
```

### –ò–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ
–î–∞–Ω—ã —Ç—Ä–∏ —Ç–æ—á–∫–∏ –ê, –í –∏ –° –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ. –°–æ—Å—Ç–∞–≤–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –ê–í–°, —Å–∏—Å—Ç–µ–º—É –ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –ø—Ä—è–º–æ–π –ê–í –∏ –ø—Ä—è–º–æ–π –ê–ú, –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω–æ–π –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –ê–í–°.

–í–∞—Ä–∏–∞–Ω—Ç N2

A(0, -1, -4), B(1, 0, -2), C(4, 4, -1).

```python
class plane:
  def __init__(this, p1, p2, p3):
    p4 = point(Symbol('x'), Symbol('y'), Symbol('z'))
    this.equation = []
    for item in tuple((p4, p2, p3)):
      this.equation.append([item.x - p1.x, item.y - p1.y, item.z - p1.z])
    this.equation = Eq(Matrix(this.equation).det(), 0)
```

```python
class line:
  def __init__(this, p1, p2):
    this.equations = Eq(
        Matrix([Symbol('x'), Symbol('y'), Symbol('z')]),
        Matrix([p1.x + Symbol('t') * (p2.x - p1.x), p1.y + Symbol('t') * (p2.y - p1.y), p1.z + Symbol('t') * (p2.z - p1.z)]),
        evaluate=False
    )
```

```python
user_data = input_data("""0 -1 -4
1 0 -2
4 4 -1""")

point1 = user_data.get_input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–∫–∏ –ê —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split(' ')
point2 = user_data.get_input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–∫–∏ B —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split(' ')
point3 = user_data.get_input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–∫–∏ C —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split(' ')

p = plane(
    point(int(point1[0]), int(point1[1]), int(point1[2])),
    point(int(point2[0]), int(point2[1]), int(point2[2])),
    point(int(point3[0]), int(point3[1]), int(point3[2]))
)
display(p.equation)

l = line(
    point(int(point1[0]), int(point1[1]), int(point1[2])),
    point(int(point2[0]), int(point2[1]), int(point2[2]))
)
display(l.equations)

koeffs = sympy.Poly(p.equation, Symbol('x'), Symbol('y'), Symbol('z')).coeffs()
l = line(
    point(int(point1[0]), int(point1[1]), int(point1[2])),
    point(int(point1[0]) + koeffs[0], int(point1[1]) + koeffs[1], int(point1[2]) + koeffs[2]),
)
display(l.equations)
```
