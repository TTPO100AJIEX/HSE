```python
import sympy
from sympy import  Matrix, symbols
from sympy import sin as Sin
from sympy import cos as Cos
from sympy import pi as Pi
from sympy.vector import CoordSys3D, Vector, matrix_to_vector, AxisOrienter, express
from IPython.display import Math, Pretty
```

```python
class input_data:
  def __init__(this, data):
    this.input_data = data.split("\n")
    this.last_used_index = -1
  def get_input(this, question):
    this.last_used_index = this.last_used_index + 1
    if (this.last_used_index >= len(this.input_data)):
      this.input_data.append(input(question))
    else:
      print(f"""{question}{this.input_data[this.last_used_index]}""")
    return(this.input_data[this.last_used_index])

  def read_coord_system(this):
    return(CoordSys3D(this.get_input("Название системы координат: ")))
  def read_vec3(this, coord_system):
    coords = this.get_input("Координаты вектора (x, y, z): ")[1:-1].split(', ')
    for i in range(3):
      coords[i] = sympy.parse_expr(coords[i])
    return(coords[0] * coord_system.i + coords[1] * coord_system.j + coords[2] * coord_system.k)

  def read_expr(this):
    return(sympy.parse_expr(this.get_input("Выражение: ")))
  def read_matrix(this, height, width):
    data = []
    for i in range(height):
      line = this.get_input(f"Строка {i + 1} матрицы: ").split(' ')
      for i in range(len(line)):
        line[i] = sympy.parse_expr(line[i])
      data.append(line)
    return(sympy.Matrix(data))
```

### Задание 2
Ввести безымянную систему координат M и символы $p$ и $q$.
Вычислить $((v + u)\cdot(v - u)) + (((2v + u)\times(v - 3u))\cdot(2v - 3u)) - ((3v - 2u)\cdot((5v - u)\times(- u)))$, $u(q, -p, q + p)$, $v(p, 2p - q, 1)$. Скалярное произведение обозначено $\cdot$, векторное $\times$.
Использовать  & и ^, скобки расставить как в условии, поскольку приоритет у & и ^ ниже, чем у + и -. Упростить полученное выражение.

```python
user_data = input_data("""
(q, -p, q + p)
(p, 2 * p - q, 1)""")

M = user_data.read_coord_system()
u = user_data.read_vec3(M)
v = user_data.read_vec3(M)

res = ((v + u) & (v - u)) + (((2 * v + u) ^ (v - 3 * u)) & (2 * v - 3 * u)) - ((3 * v - 2 * u) & ((5 * v - u) ^ (-u)))

display(sympy.simplify(res))
```
