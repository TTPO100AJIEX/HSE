###  Задание 3.
Упростите выражение
$$
\frac{(u+v)^2-4uv}{u-v}+2v-1
$$

```python
u, v = symbols('u v')
expr = ((u + v) ** 2 - 4 * u * v) / (u - v) + 2 * v - 1
simplify(expr)
```

###  Задание 5.
Разложите на множители
$2u^4+3u^3v-2u^2v^2-2u^2v-3uv^2+2v^3$

```python
u, v = symbols('u v')
terms = [2 * (u ** 4),
         3 * (u ** 3) * v,
         -2 * (u ** 2) * (v ** 2),
         -2 * (u ** 2) * v,
         -3 * u * (v ** 2),
         2 * (v ** 3)]
from numpy import sum
expr = sum(terms)
factor(expr)
```

###  Задание 6.
Сгруппируйте выражение из Задания 5 по степеням $u$.

```python
collect(expr, u)
```

### Индивидуальное задание
Составить матрицу левой части СЛАУ, вектор-столбец переменных, вектор-столбец правой части, вектор-столбец $X_0$. Показать, что $X_0$ является решением данной СЛАУ.

N2
\begin{align*}
        \left\{\begin{matrix}3 x_{1} - 8 x_{2} + 8 x_{3} - 9 x_{4} + 9 x_{5} = -120
\\- 6 x_{1} + 9 x_{2} - 3 x_{3} - 3 x_{4} + 6 x_{5} = 9
        \\- 7 x_{1} + 2 x_{2} - 7 x_{3} + 3 x_{4} - 6 x_{5} = 126
\end{matrix}\right.\qquad X_0=\left[\begin{matrix}-6 & -3 & -9 & 3 & -3\end{matrix}\right]
 \end{align*}
 

```python
import re
import sys

class sole:
  equations = []
  results = []
###############################################
  def parse_equation(this, equation):
    coeffs = re.split('x[^-+=]', equation.replace(" ", "").replace('=', ''))
    this.equations.append([])
    for i in range(len(coeffs)):
      try:
        if (i == len(coeffs) - 1):
          this.results.append(int(coeffs[i]))
        else:
          this.equations[len(this.equations) - 1].append(int(coeffs[i]))
      except:
        print("\033[91mВведённые данные некорректны")
        sys.exit(0)
###############################################
  def check_colution(this, solution):
    try: 
      coeffs = Matrix(this.equations)
    except:
      print("\033[91mВведённые данные некорректны")
      sys.exit(0)
    print("\033[95mМатрица левой части СЛАУ:", end = "")
    display(coeffs)
    
    print("\033[95mВектор-столбец переменных:", end = "")
    display(Matrix(symbols(f"""x1:{len(solution) + 1}""")))
    solutions = Matrix(solution)
    print("\033[95mВектор-столбец переменных с соответствующими значениям:", end = "")
    display(solutions)

    try: 
      result = coeffs * solutions
    except:
      print("\033[91mВведённые данные некорректны")
      sys.exit(0)
    print("\033[95mПроизведение:", end = "")
    display(result)
    
    expected_result = Matrix(this.results)
    print("\033[95mОждидаемый результат:", end = "")
    display(expected_result)

    if (result == expected_result):
      print("\033[92mРешение подходит:")
    else:
      print("\033[91mРешение не подходит:")

    for j in range(len(this.equations)):
      for i in range(len(solution)):
        print(f"""\033[94m({this.equations[j][i]})\033[95m*\033[94m({solution[i]})""", end = " ")
        if (i != len(solution) - 1):
          print("\033[95m+", end = " ")
      print(f"""\033[95m= \033[94m{result[j]}""", end = "")
      if (result[j] == this.results[j]):
        print(f"""\033[92m = {this.results[j]}""")
      else:
        print(f"""\033[91m ≠ {this.results[j]}""")
###############################################


try: 
  equations_amount = int(input("Введите количество уравнений:"))
except:
  print("\033[91mВведённые данные некорректны")
  sys.exit(0)

system = sole()

for i in range(equations_amount):
  system.parse_equation(input("Введите уравнение (не опускайте переменные с нулевыми коэффициентами и не меняйте порядок переменных):"))
#3x1 - 8x2 + 8x3 - 9x4 + 9x5 = -120
#-6x1 + 9x2 - 3x3 - 3x4 + 6x5 = 9
#-7x1 + 2x2 - 7x3 + 3x4 - 6x5 = 126

solution = input("Введите предполагаемые значения переменных через пробел:").split(' ')
#-6 -3 -9 3 -3
  
system.check_colution(solution)
```
