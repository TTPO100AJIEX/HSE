### Задание 1.
Проверить по теореме Кронекера-Капелли совместность СЛАУ $AX = b$, если
\begin{align*}
a)\ A = \left(
\begin{matrix}
-1 & -3 & 5 & 2\\
2 & -1 & 4 & 7\\
3 & 4 & -1 & 0\\
5 & 0 & 1 & -3
\end{matrix}
\right), \quad 
b = \left(
\begin{matrix}
14\\ 56\\ 5\\ -6
\end{matrix}
\right), \quad 
b)\ A = \left(
\begin{matrix}
1 & 3 & 5 & 2\\
2 & -1 & 4 & 1\\
3 & 4 & -1 & 0\\
5 & 3 & 3 & 1
\end{matrix}
\right), \quad 
b = \left(
\begin{matrix}
24\\ 20\\ 5\\ 5
\end{matrix}
\right), \\
c)\ A = \left(
\begin{matrix}
1 & 3 & 5 & 2\\
2 & -1 & 4 & 1\\
3 & 2 & 9 & 3\\
1 & -4 & -1 & -1
\end{matrix}
\right), \quad 
b = \left(
\begin{matrix}
24\\ 20\\ 44\\ -4
\end{matrix}
\right)
\end{align*}

```python
class SOLE:
  def __init__(this, variables_amount):
    this.variables = [None] * variables_amount

    this.left_side = []
    this.right_side = []

    this.augmented = []

    this.input_variables()
    this.input_equations()
  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = Symbol(f"""x{i}""")
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("Введите коэффициенты уравнения через один пробел: ");
      koeffs = equation.split(' ')
      try:
        this.left_side.append([])
        for i in range(len(this.variables)):
          this.left_side[len(this.left_side) - 1].append(parse_expr(koeffs[i]))
        this.right_side.append(koeffs[len(koeffs) - 1])
      except:
        print("\033[91mВведённые данные некорректны")
        sys.exit(0)
    this.left_side = Matrix(this.left_side)
    this.right_side = Matrix(this.right_side)
    this.augmented = this.left_side.row_join(this.right_side)

  def check_sole(this):
    if (this.left_side.rank() == this.augmented.rank()):
      return(True)
    else:
      return(False)
  def check_unique_solution(this):
    if (this.check_sole() and this.left_side.rank() == len(this.variables)):
      return(True)
    else:
      return(False)
```

### Задание 3
Для однородных СЛАУ, соответствующих СЛАУ из Задания 1 установить, есть ли у однородной СЛАУ нетривиальное решение. 

```python
class homogeneous_SOLE:
  def __init__(this, variables_amount):
    this.variables = [None] * variables_amount

    this.left_side = []
    this.right_side = []

    this.augmented = []

    this.input_variables()
    this.input_equations()
  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = Symbol(f"""x{i}""")
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("Введите коэффициенты уравнения через один пробел: ");
      koeffs = equation.split(' ')
      try:
        this.left_side.append([])
        for i in range(len(this.variables)):
          this.left_side[len(this.left_side) - 1].append(parse_expr(koeffs[i]))
        this.right_side.append(0)
      except:
        print("\033[91mВведённые данные некорректны")
        sys.exit(0)
    this.left_side = Matrix(this.left_side)
    this.right_side = Matrix(this.right_side)

  def check_untrivial_solution(this):
    if (this.left_side.rank() < len(this.variables)):
      return(True)
    else:
      return(False)
  def find_untrivial_solution(this):
    return(this.left_side.nullspace())
  def get_common_solution(this):
    solutions = this.find_untrivial_solution()
    X = this.right_side
    for i, solution in enumerate(solutions):
      X += solution * Symbol(f"""c{i + 1}""")
    return(X)
  def get_common_solution_linsolve(this):
    return(linsolve((this.left_side, this.right_side), symbols(f"""c1:{len(this.variables) + 1}""")))
```

### Индивидуальное задание.
Дана СЛАУ $AX = b$. Проверить совместность по теореме Кронекера-Капелли. Если СЛАУ совместна, проверить единственность решения.
Для соответствующей однородной СЛАУ проверить существование нетривиального решения. В случае, если оно существует, 
найти размерность пространства решений и составить ФСР и общее решение однородной СЛАУ.

N2
\begin{align*}
 A = \left[\begin{matrix}5 & 1 & -3 & -5 & 8\\-4 & 7 & -7 & 1 & -5\\4 & 32 & -40 & -16 & 12\\36 & -24 & 16 & -24 & 52\\9 & -5 & 2 & -7 & 7\end{matrix}\right],
\ b = \left[\begin{matrix}6\\81\\348\\-300\\-31\end{matrix}\right]. 
 \end{align*}

```python
class SOLE:
  def __init__(this, variables_amount):
    this.variables = [None] * variables_amount

    this.left_side = []
    this.right_side = []

    this.augmented = []

    this.input_variables()
    this.input_equations()
  def input_variables(this):
    for i in range(len(this.variables)):
      this.variables[i] = Symbol(f"""x{i}""")
  def input_equations(this):
    for i in range(len(this.variables)):
      equation = user_data.get_input("Введите коэффициенты уравнения через один пробел: ");
      koeffs = equation.split(' ')
      try:
        this.left_side.append([])
        for i in range(len(this.variables)):
          this.left_side[len(this.left_side) - 1].append(parse_expr(koeffs[i]))
        this.right_side.append(koeffs[len(koeffs) - 1])
      except:
        print("\033[91mВведённые данные некорректны")
        sys.exit(0)
    this.left_side = Matrix(this.left_side)
    this.right_side = Matrix(this.right_side)
    this.augmented = this.left_side.row_join(this.right_side)

  def check_sole(this):
    if (this.left_side.rank() == this.augmented.rank()):
      return(True)
    else:
      return(False)
  def check_unique_solution(this):
    if (this.check_sole() and this.left_side.rank() == len(this.variables)):
      return(True)
    else:
      return(False)

  def make_homogeneous(this):
    this.right_side = zeros(len(this.variables), 1)

  def check_untrivial_solution(this):
    if (this.left_side.rank() < len(this.variables)):
      return(True)
    else:
      return(False)
  def find_untrivial_solution(this):
    return(this.left_side.nullspace())
  def get_common_solution(this):
    solutions = this.find_untrivial_solution()
    X = this.right_side
    for i, solution in enumerate(solutions):
      X += solution * Symbol(f"""c{i + 1}""")
    return(X)
```
