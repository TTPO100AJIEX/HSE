```python
import sympy
from sympy import Symbol, symbols, S, Function, integrate, Integral, diff, solve
```

```python
class input_data:
  def __init__(this, data):
    this.input_data = data.split("\n")
    this.last_used_index = -1
  def get_input(this, question):
    this.last_used_index = this.last_used_index + 1
    if (this.last_used_index >= len(this.input_data)):
      this.input_data.append(input(question))
    else:
      print(f"""{question}{this.input_data[this.last_used_index]}""")
    return(this.input_data[this.last_used_index])


  def read_expression(this, question):
    expr = this.get_input(question)
    return(sympy.parse_expr(expr))

  def read_number(this, question):
    val = this.get_input(question)
    if (val == "inf"):
      return(sympy.oo)
    if (val == "-inf"):
      return(-sympy.oo)
    return((int)(val))
    
  def read_symbol(this, question):
    val = this.get_input(question)
    return(sympy.Symbol(val))
```

```python
class integrator:
  def __init__(this, func):
    this.function = func
    if (len([*this.function.free_symbols]) != 1):
      this.variable_needed = True
      return
    this.variable_needed = False
    this.variable = [*this.function.free_symbols][0]
  def get_indefinite_integral(this):
    return(Integral(this.function, this.variable))
  def get_definite_integral(this, lim1, lim2):
    return(Integral(this.function, (this.variable, lim1, lim2)))
    
  def find_definite_integral(this, lim1, lim2):
    if (this.variable_needed):
      return(None)
    return(this.function.integrate((this.variable, lim1, lim2)))

  def find_indefinite_integral(this):
    return(this.function.integrate(this.variable) + sympy.Symbol('C'))
    
  def change_variable(this, I, to_change, variable):
    return(I.transform(to_change, variable))
```

### Задача 1.
Вычислить определенные интегралы:
$$
1)\ \int_2^3 \frac{1}{\sqrt{x-2}},\ 2) \int_1^6 \frac{1}{(x-1)^3},\ 3)\ \int_2^\infty \frac{1}{(x-1)^3} 
$$

```python
user_data = input_data("""1 / sqrt(x - 2)
2
3
1 / ((x - 1) ** 3)
1
6
1 / ((x - 1) ** 3)
2
inf""")

for i in range(3):
  integral = integrator(user_data.read_expression(f"{i + 1}) function = "))
  if (integral.variable_needed):
    integral.set_variable(user_data.read_symbol(f"{i + 1}) variable = "))
  display(integral.find_definite_integral(user_data.read_number(f"{i + 1}) lim1 = "), user_data.read_number(f"{i + 1}) lim2 = ")))
```

### Задача 3.
Найти первообразную функции $\frac{x-3}{(x+1)^2}$, график которой проходит через начало координат.

```python
user_data = input_data("""(x - 3) / ((x + 1) ** 2)""")
integral = integrator(user_data.read_expression(f"function = "))
F = integral.find_indefinite_integral()

C_values = [*sympy.solveset(F.subs({integral.variable: 0}))]
for i in range(len(C_values)):
  display(F.subs({sympy.Symbol('C'): C_values[i]}))
```

### Задача 4.
Описать функцию, вычисляющую неопределенный интеграл от $\frac{1}{\sqrt{x^2+a}}$ по формуле "длинного логарифма":
$$
\int\frac{1}{\sqrt{x^2+a}} = \ln|x + \sqrt{x^2 + a}| + C,\quad a \neq 0. 
$$
Функция возвращает $\ln|x| + C$ при $a = 0$. Аргументы функции - переменная интегрирования $x$ и параметр $a$.

Вызвать функцию при $a = 0, 1, 4$. Сравнить с результатом integrate.

Проверить дифференцированием, при необходимости использовать simplyfy для того, чтобы упростить выражение.

Рекомендется создавать положительные символы, чтобы не было неоднозначности.

```python
def long_log(x, a):
  if (a == 0):
    return(sympy.log(abs(x)) + sympy.Symbol('C'))
  return(sympy.log(abs(x + sympy.sqrt(x ** 2 + a))) + sympy.Symbol('C'))

x = sympy.Symbol('x', positive=True)
a = sympy.Symbol('a', positive=True)

f = 1 / sympy.sqrt(x ** 2 + a)
A_s = [0, 1, 4]
for i in range(len(A_s)):
  print("\033[95m--------------------------------------\033[0m")
  val1 = long_log(x, A_s[i])
  val2 = integrator(f.subs({a: A_s[i]})).find_indefinite_integral()
  print("\033[95mФункция:\033[0m")
  display(val1)
  display(sympy.simplify(val1.diff(x)))
  print("\033[95mIntegrate:\033[0m")
  display(val2)
  display(sympy.simplify(val2.diff(x)))
```

### Задача 5.
Провести замену переменной $y = x^2$ в интеграле $\int x e^{x^2}\,dx$.

```python
user_data = input_data("""x * E ** (x ** 2)""")
integral = integrator(user_data.read_expression(f"function = "))
I1 = integral.get_indefinite_integral()
display(sympy.Eq(I1, integral.change_variable(I1, sympy.Symbol('x') ** 2, sympy.Symbol('y'))))
```
