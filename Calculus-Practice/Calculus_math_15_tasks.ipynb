### Задание 1
Найти точки экстремума:

1) $z = x^2 + xy + y^2 - 2x - y$

2) $z = x^3y^2(6 - x - y)$, $(x, y > 0)$

3) $z = \frac{8}{x} + \frac{x}{y} + y$

4) $z = \frac{1 + x - y}{\sqrt{1 + x^2 + y^2}}$
###### Указание.
Вначале найти стационарные точки. Составить определитель из вторых производных в произвольной точке. Подставляя найденные (вещественные) стационарные точки, решить вопрос о наличии экстремума.

```python
def find_extr(f, x, y):
  stat_points1 = [*nonlinsolve([f.diff(x), f.diff(y)], [x, y])]
  stat_points = []
  for item in stat_points1:
    if (item[0].is_real and item[1].is_real):
      stat_points.append(item)
  Delta = Matrix([[f.diff(x, 2), f.diff(x, y)], [f.diff(x, y), f.diff(y, 2)]]).det()
  for point in stat_points:
    if (Delta.subs({x: point[0], y: point[1]}) < 0):
      continue
    if (Delta.subs({x: point[0], y: point[1]}) > 0):
      A = f.diff(x, 2).subs({x: point[0], y: point[1]})
      B = f.diff(y, 2).subs({x: point[0], y: point[1]})
      if A > 0 or B > 0:
        display('minimum', point)
      elif A < 0 or B < 0:
        display('maximum', point)
      display(f.subs({x: point[0], y: point[1]}))
      continue

x = sympy.Symbol('x')
y = sympy.Symbol('y')
f = x ** 2 + x * y + y ** 2 - 2 * x - y
display(f)
find_extr(f, x, y)
print("\n\n")

x = sympy.Symbol('x', positive=True)
y = sympy.Symbol('y', positive=True)
f = (x ** 3) * (y ** 2) * (6 - x - y)
display(f)
find_extr(f, x, y)
print("\n\n")

x = sympy.Symbol('x')
y = sympy.Symbol('y')
f = 8 / x + x / y + y
display(f)
find_extr(f, x, y)
print("\n\n")

f = (1 + x - y) / sympy.sqrt(1 + x ** 2 + y ** 2)
display(f)
find_extr(f, x, y)
```

### Задание 2.
С помощью метода Лагранжа найти условный экстремум функции
$z = x + 2y$ при условии $x^2 + y^2 = 5$.

```python
x, y, lam = symbols('x y lamda')
varrs = [x, y, lam]
f = x + 2 * y
restriction = x**2 + y**2 - 5
L = f + lam*restriction
stat_points = nonlinsolve([L.diff(var) for var in varrs], varrs)

dx, dy, dx2, dy2 = symbols('dx dy dx^2 dy^2')
d2L = L.diff(x,2)*dx2 + 2*L.diff(x,y)*dx*dy + L.diff(y,2)*dy2
display(factor(d2L))

for item in stat_points:
  if (item[2] < 0):
    display('maximum', (item[0], item[1]))
    display(f.subs({x: item[0], y: item[1]}))
  if (item[2] > 0):
    display('minimum', (item[0], item[1]))
    display(f.subs({x: item[0], y: item[1]}))
```

### Задание 3.
С помощью метода Лагранжа найти условный экстремум функции
$z=x^2+y^2$ при условии $\frac{x}{2}+\frac{y}{3}=1$.

```python
x, y, lam = symbols('x y lamda')
varrs = [x, y, lam]
f = x**2 + y**2
restriction = x / 2 + y / 3 - 1
L = f + lam*restriction
stat_points = nonlinsolve([L.diff(var) for var in varrs], varrs)
dx, dy, dx2, dy2 = symbols('dx dy dx^2 dy^2')
d2L = L.diff(x,2)*dx2 + 2*L.diff(x,y)*dx*dy + L.diff(y,2)*dy2
display(factor(d2L))
for item in stat_points:
    display('minimum', (item[0], item[1]))
    display(f.subs({x: item[0], y: item[1]}))
```

### Задание 4.
С помощью метода Лагранжа найти экстремум функции $u = x - 2y + 2z$ при условии $x^2 + y^2 + z^2 = 9$.

```python
x, y, z, lam = symbols('x y z lamda')
varrs = [x, y, z, lam]
u = x - 2 * y + 2 * z
restriction = x ** 2 + y ** 2 + z ** 2 - 9
L = u + lam*restriction
stat_points = nonlinsolve([L.diff(var)for var in varrs], varrs)
display(stat_points)
dx, dy, dz, dx2, dy2, dz2 = symbols('dx dy dz dx^2 dy^2 dz^2')
d2L = 0
diffs = {x: (dx, dx2), y: (dy, dy2), z: (dz, dz2)}
for i, var in enumerate(varrs[:-1]):
    d2L += L.diff(var, 2)*diffs[var][1]
    for j in range(i + 1, 3):
        d2L += 2*L.diff(var, varrs[j])*diffs[var][0]*diffs[varrs[j]][0]
display(factor(d2L))
for item in stat_points:
  if (item[3] > 0):
    display('minimum', (item[0], item[1], item[2]))
    display(u.subs({x: item[0], y: item[1], z: item[2]}))
  if (item[3] < 0):
    display('maximum', (item[0], item[1], item[2]))
    display(u.subs({x: item[0], y: item[1], z: item[2]}))
```
