### Задание 1
Вычислить площадь фигуры, ограниченной графиками функций $f = 2 + x^2$ и  $g = 2 - x^3$. Построить графики, найти точки пересечения, затем вычислить соответствующие интегралы.

```python
x = sympy.Symbol('x')
f = 2 + x ** 2
g = 2 - x **  3

p1 = plot(f, show=False, xlim=(-3, 0), ylim=(1, 4))
p1.append(plot(g, show=False, xlim=(-3, 0), ylim=(1, 4))[0])
p1.show()

X = [*sympy.solveset(sympy.Eq(f, g))]
print("Точки пересечения:")
for i in range(len(X)):
  print(f"({X[i]}, {f.subs(x, X[i])})")

print("Площадь:")
display((f - g).integrate((x, X[0], X[1])))
```

### Задача 2.
Вычислить площадь фигуры, ограниченной графиком функции $x = \cos^3(t)$, $y = \sin^3(t)$, $0 \le t \le 2\pi$.

```python
t = sympy.Symbol('t')

x = sympy.cos(t) ** 3
y = sympy.sin(t) ** 3
#plot_parametric((x, y), (t, 0, 2 * Pi))

display((x * y.diff(t)).integrate((t, 0, 2 * Pi)))
```

### Задача 3.
Вычислить площадь фигуры, ограниченной графиком функции $r = \cos(2\phi)$, $-\frac{\pi}{4} \le \phi \le \frac{\pi}{4}$. 

```python
phi = S('phi')
r = sympy.cos(2 * phi)
display(S(1) / 2 * integrate(r * r, (phi, -Pi / 4, Pi / 4)))
```

### Задача 4.
Вычислить длину кривой, заданной уравнением $y^2 = x^3$ при $x$ от 1 до 3.

```python
x = sympy.Symbol('x')
y = sympy.sqrt(x ** 3)
sympy.simplify(2 * ((1 + (y.diff(x) ** 2)) ** S.Half).integrate((x, 1, 3)))
```

### Задача 6.
Вычислить длину кривой, заданной в полярных координатах
$r = \phi$, $\phi$ от 0 до $2\pi$. Результат вычислить приближенно до 3 знаков после запятой.

```python
phi = S('phi')
r = phi
display(sympy.simplify(((r ** 2 + r.diff(phi) ** 2) ** S.Half).integrate((phi, 0, 2 * Pi))).evalf(5))
```

### Индивидуальное задание
Вычислить площадь фигуры, ограниченной графиками функций $f = 2\sin(x)$ и  $g = \cos(2x)$ на интервале  $[0, \pi]$. Построить графики, найти точки пересечения (выделить корни на интервале с помощью intersect и Interval), затем вычислить соответствующие интегралы.

Вариант 2

$$f(x)=\cos{\left(4 x \right)}$$
$$g(x)=2 \cos{\left(2 x \right)}$$

на интервале $[0.2 \pi;0.8 \pi]$


```python
x = sympy.Symbol('x')
f = sympy.cos(4 * x)
g = 2 * sympy.cos(2 * x)

p1 = plot(f, show=False, xlim=(0, Pi), ylim=(-3, 3))
p1.append(plot(g, show=False, xlim=(0, Pi), ylim=(-3, 3))[0])
p1.show()

X = sympy.solveset(sympy.Eq(f, g), domain=sympy.Reals)
Int = Interval(0.2 * Pi, 0.8 * Pi)
X = [*Int.intersect(X)]
print("Точки пересечения:")
for i in range(len(X)):
  print(f"({X[i].evalf()}, {f.subs(x, X[i]).evalf()})")

print("Площадь:")
res = (g - f).integrate((x, X[0], X[1]))

display(sympy.simplify(res))
display(res.evalf())
```
