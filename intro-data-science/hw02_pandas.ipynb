```python
print('Q: Какая доля студентов сдала экзамен по математике (passmark >= 50)?')
column = df['math score']
print(f'A: {column[column >= passmark].count() / column.count()}')

print('Q: Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?')
column = df[df['test preparation'] != 'none']['math score']
print(f'A: {column[column >= passmark].count() / column.count()}')

print('Q: Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?')
column = df[(df['gender'] == 'female') & (df['test preparation'] == 'none')]['math score']
print(f'A: {column[column < passmark].count() / column.count()}')
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
%%time
sum = dict()
count = dict()
for index, row in df.iterrows():
    group = row['race/ethnicity']
    sum[group] = sum.get(group, 0) + row['reading score']
    count[group] = count.get(group, 0) + 1
for key in sorted(count.keys()):
  print(f'{key}: {sum[key] / count[key]}')
```

```python
%%time
ans = dict()
for index, row in df.iterrows():
    group = row['education']
    if (group in ans):
      ans[group] = min(ans[group], row['writing score'])
    else:
      ans[group] = row['writing score']
for key in sorted(ans.keys()):
  print(f'{key}: {ans[key]}')
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
def addPassColumn(srcColumn, destColumn):
  df[destColumn] = (df[srcColumn] >= passmark).map({ False: 'F', True: 'P' })
  return df[destColumn].value_counts().to_dict()
print(f'Math: {addPassColumn("math score", "Math_PassStatus")}')
print(f'Reading: {addPassColumn("reading score", "Read_PassStatus")}')
print(f'Writing: {addPassColumn("writing score", "Write_PassStatus")}')
df
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
  if (average_mark > 90):
    return 'A'
  if (average_mark >= 80):
    return 'B'
  if (average_mark >= 70):
    return 'C'
  if (average_mark >= 60):
    return 'D'
  if (average_mark >= 50):
    return 'E'
  return 'F'

df['Grade'] = ((df['math score'] + df['reading score'] + df['writing score']) / 3).map(GetGrade)
df
```
